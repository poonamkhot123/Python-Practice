# List Slicing
Slicing allows you to access a range of elements in a list.You can specify the start and stop indices,and Python returns a new list containing the specified elements.

**Syntax: list_name[Start:stop:step]**

```python
#example:
numbers = [10,20,30,40,50,60]

#slice from index 1 to 3
print(numbers[1:4])

#slice from start to index 2
print(numbers[:3])

#slice all alternate elements
print(numbers[::2])

#slice with negative indices.
print(numbers[-4:-1])

#Reverse list.
print(numbers[::-1])

#Last three elements
print(numbers[3:])
```

* **unique elements** : No duplicates allowed. Each element must be distinct.
* **mutable** : you can add or remove elements after creation.

```python
# Create a set in python
# There are two primary ways to create a set in python:

# 1.Using curly Braces{}
my_set = {1,2,3,4,5}
print(my_set)

#2.Using the set() Constrctor
my_set = set([1,2,3,4,5])
print(my_set)

print()

# note: An empty set cannot be created using{} as it creates an empty dictomnary.
empty_set = set()
print(empty_set)
```

# Set Methods



```python
# 1. Union: combines elements from two sets,removing duplicates.
set_a = {1,2,3}
set_b = {3,4,5}
union_set = set_a.union(set_b)
print(union_set)
```

```python
# 2. Intersection : includes only elements peresent in both sets.
set_a = {1,2,3}
set_b = {3,4,5}
intersection_set = set_a.intersection(set_b)
print(intersection_set)



# intersection alternative
inter_set2 = set_a & set_b
print(inter_set2)
```

```python
# 3.Difference - element present in first set only but not in second set.
set1 = {1,2,3,4}
set2 = {3,4,5}
diff_set = set1.difference(set2)
print(diff_set)

# Difference alternative
diff_set2 = set1 - set2
print(diff_set2)
```

```python
# 4.Symmetric difference : element in either set but not in both
set1 ={1,2,3,4}
set2 = {3,4,5,6}
sdiff_set = set1.symmetric_difference(set2)
print(sdiff_set)

# symm Diff alternative
sdiff_set2 = set1 ^ set2
print(sdiff_set2)
```

* Lambda functions are often used for short, simple operations without needing to define a full function using def.

* Suntax:

* Arguements: input to the function
* Expression : A single statement or operation that the lambda function

```python
# lambda function examples
def add(x,y):
    return x+y
print(add(3,5))

# Add two numbers using a lambda function
add = lambda x,y:x+y
print(add(3,5))

# Custom sorting - sort a list of tuples by the second elements.
data = [(1,'b'),(3,'a'),(2,'c')]
sorted_data = sorted(data,key = lambda x:x[1])
print(sorted_data)

```

```python
# Q.2 find the most Frequent Element in a list
numbers = [1,2,2,3,3,3,4]
def most_freq(lst):
  max_count = 0
  most_freq = None
  for item in lst:
    count = lst.count(item)
    if count > max_count:
      max_count = count
      most_freq = item
    return most_freq
print(most_freq (numbers))

```

```python
# Q.6 Find the Most Frequent value in a dictionary
data = {'a':1,'b':2,'c':1,'d':3,'e':1}
def most_freq(dct):
  frequency = {}
  for value in dct.values():
    if value not in frequency:
      frequency[value] = 0
    frequency[value] += 1
  max_value = max(frequency,key=frequency.get)
  return max_value
print(most_freq(data))
```

```python
# Q7 Merge Dictionaries with Summation.
dict1 = {'a':10,'b':20,'c':30}
dict2 = {'b':15,'c':35,'d':25}
def merge_dict(dict1,dict2):
  result = dict1.copy()
  for key,value in dict2.items():
    if key in result:
      result[key] += value
    else:
      result[key] = value
    return result
print(merge_dict(dict1,dict2))
```

# **OOPs in Python**
* **Two ways of programming in Python:**

* 1) Procedural programming ,
* 2) OOPs

**OOPs:object oriented programming**
Away of organizing code by creating "blueprints" (called classes) to represent real- world things like a student,car,or house. These blueprints help you create objects (individual examples of these things) and define theit behavior.

* **Class:** Aclass is bluprint or template for creating objects.
it defines the properties(attributes) and actions/behaviors (methods) type will have.

* **Object:** An object is a specific of a class. It has actual data based on the blueprint defined by the class.

# why OOPs ?
* models real-world problems
* code reuasability
* easier maintenance
* encapsulation
* flexibility and scalability

```python
# using OOPs - creating student records

# class - blueprint or template
class Student:
  name = 'Madhav'
  grade = 10

# object - instance of class
student1 = Student()
print(student1.name)
```

```python
class Student():
  def __init__(self,name,grade,percentage):   # method
    self.name = name                          # attribute
    self.grade = grade                        # attribute
    self.percentage = percentage              # attribute

  def student_details(self):
    print(f"{self.name} is in class {self.grade},with {self.percentage}%")

# object - instance of class
student1 = Student('madhav',11,96)
student2 = Student('Vishakha',12,99)

student1.student_details()
student2.student_details()

# if we want information with key value pair
print(student1.__dict__)

# modify object property
print(student1.percentage)
student1.percentage = 100 #modify
print(student1.percentage)

# delete object
del student1
print(student1)
```

```python
# Abstraction - hiding unnecessary details from users through class,methods

class Student():
  def __init__(self,name,grade,percentage):   # method
    self.name = name                          # attribute
    self.grade = grade                        # attribute
    self.percentage = percentage


  def student_details(self):   #method -abstraction
    print(f"{self.name} is in class {self.grade},with {self.percentage+2}%")
    # Hidden from users.

# object - instance of class
student1 = Student('madhav',11,96)
student2 = Student('Vishakha',12,97)

student1.student_details()
student2.student_details()
```

```python
# Encapsulation: Restrict access to certain attributes or methods to protect data and enforce controlled access

class Student():
  def __init__(self,name,grade,percentage):   # method
    self.name = name                          # attribute
    self.grade = grade                        # attribute
    self.__percentage = percentage # double underscore limits access

  def get_percentage(self):
    return self.__percentage

  def student_details(self):
    print(f"{self.name} is in class {self.grade},with {self.percentage}%")
    # Hidden from users.

# object - instance of class
student1 = Student('madhav',11,96)
student2 = Student('Vishakha',12,97)

print(student2.get_percentage)
```
