# Sequence Sum
* 1/1! + 2/2! + 3/3!+.....

```python
n  = int(input("ENter a number:"))
result = 0
fact = 1

for i in range(1,n+1):
  fact = fact *i
  result = result + i/fact
print(result)
```

# Pattern 2
* 1
* 121
* 12321
* 1234321

```python
n = int(input("ENter a number of rows:"))

for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end = " ")
  print()
```

```python
n = int(input("ENter a number of rows:"))

for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end = " ")
  for k in range(i-1,0,-1):
    print(k,end = " ")
  print()
```

# Loop control Statement
* break
* Continue
* pass



```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
lower = int(input("Enter a Number:"))
upper = int(input("Enter a Number:"))
for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break
  else:
    print(i)
```

```python
# Continue
for i in range(1,10):
  if i == 5:
    continue
  print(i)

```

```python
# Pass
for i in range(1,10):
  pass
```
